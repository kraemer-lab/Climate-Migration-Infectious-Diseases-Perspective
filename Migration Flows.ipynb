```python
#Program to process migration flows data into a datastructure
from iso3166 import countries
import pycountry_convert as pc
fp = open('data/migration-flows.csv', 'r')
lines = fp.readlines()
count = 0
column_map_from= {}
column_map_to = {}
uniqueCountries = set()
noCountries = 0
ignorelist = ['Bonaire Sint Eustatius and Saba', 'Caribbean', 'Channel Islands', 'Curacao',
              'Democratic Republic of Congo', 'Faeroe Islands', 'Polynesia', 'Reunion', 'Saint Barthlemy', 
              'Saint Helena','Timor-Leste','Vatican','Western Sahara']
fw = open('data/migration_data.csv', 'w')
for line in lines:
    count += 1
    cols = line.strip().split(",")
    if count == 1:
        continue
    year = int(cols[0])
    country = cols[1]
    if year != 2020:
        continue
    if country not in uniqueCountries:
        noCountries += 1
        uniqueCountries.add(country)
    emID = 2+2*(noCountries-1)
    imID = 2+2*(noCountries-1)+1
    if cols[emID] == '':
        continue
    if country in ignorelist:
        #print(country, "ignored")
        continue
    noE = -1*int(cols[emID])
    noI = int(cols[imID])
    netM = noI - noE #Net migration is the difference of number of immigrants and number of emigrants.
    country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")
    #print(country_code)
    continent_code = pc.country_alpha2_to_continent_code(country_code)
    continent_name = pc.convert_continent_code_to_continent_name(continent_code)
    print(country, noE, noI, netM,continent_name)
    fw.write(country+","+str(noE)+","+str(noI)+","+str(netM)+","+continent_name+"\n")
    #print(continent_name)
fp.close()
fw.close()
```

```python
import math
fp = open('data/migration_data.csv', 'r')
fw = open('data/vulnerability.csv', 'r')
vulnerability = {}
lines = fw.readlines()
count = 0
for line in lines:
    count += 1
    cols = line.strip().split(",")
    if count == 1:
        continue
    country = cols[1].strip('\"')
    if cols[-1] == '':
        continue
    vul = float(cols[-1])
    #print(country, vul)
    vulnerability[country] = vul
fw.close()
fw = open('data/Countries Population from 1995 to 2020.csv', 'r')
population = {}
lines = fw.readlines()
count = 0
for line in lines:
    count += 1
    cols = line.strip().split(",")
    if count == 1:
        continue
    country = cols[1].strip('\"')
    if cols[-1] == '':
        continue
    pop = int(cols[2])
    #print(country, vul)
    population[country] = pop
fw.close()
lines = fp.readlines()
count = 0
fr = open('data/migration_vulnerability.csv', 'w')
print(population)
for line in lines:
    count += 1
    cols = line.strip().split(",")
    country = cols[0]
    outM = int(cols[1])
    if country not in vulnerability or country not in population:
        continue
    vul_index = vulnerability[country]
    pop = population[country]
    netM = int(cols[3])
    if netM < 0:
        netM = -1 * netM
        NetMlog = -1*math.log10(netM)
    else:
        NetMlog = math.log10(netM)
    outMbyPop = float(outM/population[country]*100)
   
    #print(line, vul_index)
    fr.write(line.strip()+","+str(vul_index)+","+str(pop)+","+str(outMbyPop)+","+str(NetMlog)+"\n")
fp.close()
fr.close()
```
